```python
para_list = []

for state in capital_dic:
    capital = capital_dic[state]
    print(capital + ", " + state)
    
    query = f"https://api.openaq.org/v2/latest?limit=1&country=US&state={state}&city={capital}"
    
    response = requests.get(query).json() 
    
    found = response["meta"]["found"]
    
    if found > 0:
        measurements = response["results"][0]["measurements"]
        pprint(measurements)

        dict_temp ={}
        dict_temp["state"] = state
        dict_temp["capital"] = capital
        
        for id in ['pm10', 'pm25', 'no2', 'so2', 'o3', 'co', 'nox', 'no']:
            dict_temp[id] = 0

        for para in range(len(measurements)):
            parameter = measurements[para]["parameter"]
            value = measurements[para]["value"]

            dict_temp[parameter] = value

        para_list.append(dict_temp)   
        print("******************* \n")
        print(dict_temp)
        print("=========================== \n")
    else:
        print("=========================== \n")

    #     break    
```
